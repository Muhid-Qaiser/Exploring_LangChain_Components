# Split HTML Elements using HTML Text Splitter

```python
from langchain_text_splitters import HTMLHeaderTextSplitter

html_string = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <title>Sample HTML Document with Headings</title>
</head>
<body>
    <h1>Main Title of the Document</h1>
    <p>This is an introductory paragraph providing an overview of the document's content.</p>

    <h2>Section 1: Introduction</h2>
    <p>This section introduces the main topics discussed in the document.</p>
    <p>Here's a line break example:<br>Notice how this text appears on a new line.</p>

    <h3>Subsection 1.1: Background</h3>
    <p>This subsection provides background information relevant to the main topics.</p>

    <h2>Section 2: Details</h2>
    <p>This section delves into the specifics of the topics introduced earlier.</p>

    <h3>Subsection 2.1: Analysis</h3>
    <p>Detailed analysis of the topics is presented here.</p>

    <h3>Subsection 2.2: Discussion</h3>
    <p>This subsection offers a discussion based on the analysis.</p>

    <h2>Section 3: Conclusion</h2>
    <p>The document concludes with a summary of the findings and final thoughts.</p>
</body>
</html>
"""

header_to_split_on = [
    ("h1", "Header 1"),
    ("h2", "Header 2"),
    ("h3", "Header 3"),
]

html_splitter = HTMLHeaderTextSplitter(header_to_split_on)
html_header_splits = html_splitter.split_text(html_string)
html_header_splits
```

## Splitting a URL using HTML Text Splitter

```python
url = "https://muhid-qaiser.medium.com/decoding-the-transformer-architecture-a-complete-guide-to-the-backbone-of-modern-ai-e554db9fe0c9"


header_to_split_on = [
    ("h1", "Header 1"),
    ("h2", "Header 2"),
    ("h3", "Header 3"),
    ("h4", "Header 4"),
]

html_splitter = HTMLHeaderTextSplitter(header_to_split_on)
html_header_splits = html_splitter.split_text_from_url(url)
html_header_splits

```
